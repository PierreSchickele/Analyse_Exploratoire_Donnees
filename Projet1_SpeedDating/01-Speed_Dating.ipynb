## Peut-on pr√©dire sa propre valeur sur le march√© du dating ?

Est-ce que les personnes peuvent pr√©dire pr√©cis√©ment leur propre valeur pre√ßue dans le march√© des rencontres ? Pour rappel :
- 3_1 : comment la personne se mesure (attr = attractivit√©, sinc = sinc√©rit√©, intel = intelligence, fun, amb = ambition)
- _o : comment le partenaire per√ßoit la personne au moment de la rencontre

On ne peut pas avoir un bon estimateur de la valeur d'une personne sur le march√© des rencontres, on choisit donc de consid√©rer la moyenne des notes de ses partenaires, avec le biais que √ßa peut engendrer, notamment si le nombre de partenaires est faible (on peut tomber sur des partenaires qui notent plus ou moins gentiment que la moyenne)

```python
df['intel_o']
```

```python
# Calcul des √©carts entre l'opinion qu'a la personne d'elle-m√™me et l'avis de son partenaire
list_attributes = ['attr', 'sinc', 'intel', 'fun', 'amb']
ecarts_predi = df.groupby('iid')[['attr3_1', 'attr_o', 'sinc3_1', 'sinc_o', 'intel3_1', 'intel_o',
                                  'fun3_1', 'fun_o', 'amb3_1', 'amb_o']].mean()
for attribute in list_attributes:
    ecarts_predi[f'diff_{attribute}'] = (ecarts_predi[f'{attribute}3_1']-ecarts_predi[f'{attribute}_o'])

ecarts_predi
```

```python
# Boxplot des diff√©rents √©carts
fig = go.Figure()
fig.add_trace(
    go.Box(
        name='Attractivit√©',
        y=ecarts_predi['diff_attr']
    )
)
fig.add_trace(
    go.Box(
        name='Sinc√©rit√©',
        y=ecarts_predi['diff_sinc']
    )
)
fig.add_trace(
    go.Box(
        name='Intelligence',
        y=ecarts_predi['diff_intel']
    )
)
fig.add_trace(
    go.Box(
        name='Fun',
        y=ecarts_predi['diff_fun']
    )
)
fig.add_trace(
    go.Box(
        name='Ambition',
        y=ecarts_predi['diff_amb']
    )
)
fig.show()
```

Pour rappel, la diff√©rence de note a √©t√© calcul√©e par individu et mesure l'√©cart entre la note donn√©e par l'individu lui-m√™me et la note donn√©e par ses partenaires.

De mani√®re g√©n√©rale, on observe une diff√©rence m√©diane positive, ce qui signifie que la note attribu√©e par la personne m√™me est sup√©rieure √† celle donn√©e par ses partenaires. Quel que soit le crit√®re, cette diff√©rence fluctue (pour les personnes dans l'√©cart inter-quartile) entre 0 et 2 points, avec une valeur m√©diane autour de 1.

On peut consid√©rer "qu'en g√©n√©ral" les individus se sur-estiment et ne pr√©disent pas correctement leur valeur sur le march√© des rencontres. Cette conclusion est toutefois √† nuancer pour la raison cit√©e pr√©c√©demment. Avec un nombre de partenaires assez faible, on peut avoir des notes plus faibles.

## Vaut-il mieux √™tre la premi√®re ou la derni√®re rencontre ?

Pour obtenir un second date, vaut-il mieux √™tre premier ou dernier ?

Pour r√©pondre √† cette question, on peut regarder la proportion de personnes accept√©es en √©tant pass√©es en premier et celle accept√©es en √©tant pass√©es en dernier.

order : valeurs extr√™mes (1 et order==round) sachant que round = nombre de personnes rencontr√©es

d√©cision : dec (1 = oui pour un second date, 0 = non)

```python
# S√©lection des colonnes utiles
df_2nd_date = df[['iid', 'round', 'order', 'dec']]
df_2nd_date.head(12)
```

```python
# Filtre sur les premi√®res et les derni√®res rencontres
mask_premier = (df_2nd_date['order'] == 1)
mask_dernier = (df_2nd_date['order'] == df_2nd_date['round'])
df_2nd_date_first = df_2nd_date[mask_premier]
df_2nd_date_last = df_2nd_date[mask_dernier]

df_2nd_date_last.head(-10)
```

```python
df_2nd_date_last['dec'].sum()
```

```python
# Histogrammes du nombre de d√©cisions (en fonction des positions)
fig = go.Figure()
fig.add_trace(
    go.Histogram(
        name='D√©cision (premiers)',
        x=df_2nd_date_first['dec'].apply(lambda x: 'oui' if x==1.0 else 'non')
    )
)
fig.add_trace(
    go.Histogram(
        name='D√©cision (derniers)',
        x=df_2nd_date_last['dec'].apply(lambda x: 'oui' if x==1.0 else 'non')
    )
)
```

Ce graphique montre que pour les premiers partenaires, il y a quasiment autant de "oui" pour une deuxi√®me rencontre que de "non". En revanche, on observe moins de "oui" pour les derniers pass√©s.

Bas√© sur ces donn√©es, on pourrait donc en d√©duire qu'il vaut mieux passer en premier qu'en dernier si on veut obtenir une deuxi√®me rencontre.
